```python
import os
import time
sid = ['1101','2330']

for s in sid:
    res = requests.get('http://mops.twse.com.tw/server-java/t164sb01?step=1&CO_ID='+s+'&SYEAR=2017&SSEASON=3&REPORT_ID=C')
    res.encoding = 'big5'
    path = os.path.join('course9',s+'.html')
    f=open(path,'w')
    f.write(res.text)
    f.close()
    
    print(s)
    
    time.sleep(10)
```

```python
dfs = []
for s in sid:
    path = os.path.join('course9',s+'.html')
    dfs.append(pd.read_html(path))
```
