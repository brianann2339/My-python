```python
import requests
url = 'http://mops.twse.com.tw/nas/t21/sii/t21sc03_108_2_0.html'
r = requests.get(url)
```

```python
from io import StringIO
import pandas as pd
r.encoding = 'big5'
```

```python
dfs = pd.read_html(StringIO(r.text))
```

```python
dfs[0]
```

```python
df = dfs[0]
df=df[list(range(0,10))]

df.columns = df[df[0] == '公司代號'].iloc[0]
df
```

```python
df = df[-pd.to_numeric(df['當月營收'],errors = 'coerce').isnull()]
```
